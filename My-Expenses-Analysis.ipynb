### B.  REGRESSION FOR THE FEATURES

#### LINEAR REGRESSION

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df["Total-income"], df["Total-Expenses"])
```

```python
# Using Slope and Intercept from the regression to plot predicted values vs oberved
def predict(x):
    return slope * x + intercept

fitline = predict(df["Total-income"])

plt.scatter(df["Total-income"], df["Total-Expenses"])
plt.plot(df["Total-income"], fitline, c="r")
plt.show()
```

```python
r_value**2
```

#### Findings
* Linear regression score of 43 percent suggests that there is a very minimal linear relationship between Income and expenses so it probably meant that increase in income didnt really mean increase in expenses, so we check out for polynomial regression below.

#### POLYNOMIAL REGRESSION

```python
import numpy as np
X =  np.array(df["Total-income"])
y = np.array(df["Total-Expenses"].astype(np.float64))

# Fit X and y to a 4th degree polynomial
p4 = np.poly1d(np.polyfit(X, y, 4))
```

```python
# Visualizing the outcome
# xp is random income generated by numpy between 20000 and 100000 which our model p4 will predict on
xp = np.arange(20000, 100000, 9)
plt.scatter(X,y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
# Evaluation
from sklearn.metrics import r2_score
R2 = r2_score(y, p4(X))
R2
```

###  Findings
A 4th degree polynomial seems to be a good fit for this dataset, It still suggests that my expenses will go up  as my income goes up but it is only 61 percent accurate with its predictions maybe because the data I gave it to learn on is not enough. But its predictions can be very slightly close to the truth.

### MULTIVARIATE REGRESSION TO PREDICT THE SPENDING HABITS WHICH COULD HAVE THE MOST IMPACT ON THE TOTAL EXPENSES GOING FORWARD

```python
from sklearn.linear_model import LinearRegression

X = df[["Food", 'Responsibility', "Water", "Transport", "Recharge", "Alms", "Miscellaneous", "Investments"]]
y = df["Total-Expenses"]

rf = LinearRegression().fit(X,y)
```

```python
rf.coef_
```

```python
columns = list(X.columns)
coeffiecients = rf.coef_
```

```python
coef_df = pd.DataFrame({"Columns": columns, "Feature_Importance": coefficients})
sorted_df = coef_df.sort_values("Feature_Importance", ascending=False)
```

```python
fig, ax = plt.subplots()
ax = ax.barh(sorted_df["Columns"], sorted_df["Feature_Importance"])
plt.gca().invert_yaxis()
plt.show()
```

#### INTERPRETATIONS FROM THE LINEAR MODEL

The model was pretty sure about money spent on water, transportation and responisbiliy but midly sure about the others.

### REGRESSION USING ENSEMBLE METHODS


#### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

X = df[["Food", 'Responsibility', "Water", "Transport", "Recharge", "Alms", "Miscellaneous", "Investments"]]
y = df["Total-Expenses"]

rf2 = RandomForestRegressor().fit(X,y)
```

```python
rf2.feature_importances_
```

```python
columns = list(X.columns)
coefficients = rf2.feature_importances_

# Map coefficients to columns and convert to Dataframe
coef_df = pd.DataFrame({"Columns": columns, "Feature_Importance": coefficients})
sorted_df = coef_df.sort_values("Feature_Importance", ascending=False)

# Visualize the feature importances
import matplotlib.style as style 
style.use("seaborn-talk")
style.use("fivethirtyeight")
fig, ax = plt.subplots()
ax = ax.barh(sorted_df["Columns"], sorted_df["Feature_Importance"])
plt.gca().invert_yaxis()
plt.show()
```

### Interpretation
Random forest regressor placed importance on money spent on water, transport and food if any siginificant change to happen in expenses.

####  USING XGBOOST FOR REGRESSION PURPOSES

```python
from xgboost import XGBRegressor
from xgboost import plot_importance
import matplotlib.style as style
style.use("seaborn-talk")
style.use("ggplot")

X = df[["Food", 'Responsibility', "Water", "Transport", "Recharge", "Alms", "Miscellaneous", "Investments"]]
y = df["Total-Expenses"]

model = XGBRegressor().fit(X, y)

# plot the feature importances
plot_importance(model);
```

### Interpretations
* Xgboost predictions makes the most sense when looking at feature importance, as it places more importance on realistic  expenditures
