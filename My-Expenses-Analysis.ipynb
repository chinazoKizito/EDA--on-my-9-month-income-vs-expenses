### (B).  REGRESSION FOR THE FEATURES

#### LINEAR REGRESSION

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df["Total-income"], df["Total-Expenses"])

# Using Slope and Intercept from the regression to plot predicted values vs oberved
def predict(x):
    return slope * x + intercept

fitline = predict(df["Total-income"])
print('Rsquared Value:',{r_value**2})

style.use("fivethirtyeight")
plt.figure(figsize=(11, 5))
plt.scatter(df["Total-income"], df["Total-Expenses"], c='b')
plt.plot(df["Total-income"], fitline, c="r")
plt.show()
```

#### Findings
* Linear regression score of 43 percent suggests that there is a very minimal linear relationship between Income and expenses 
This means that linear regression  cannot better explain the relationship between these two features,
so it probably meant that increase in income didnt really mean increase in expenses, so we check out for `polynomial regression` below.

#### POLYNOMIAL REGRESSION

```python
from sklearn.metrics import r2_score

X =  np.array(df["Total-income"])
y = np.array(df["Total-Expenses"].astype(np.float64))

# Fit X and y to a 4th degree polynomial
p4 = np.poly1d(np.polyfit(X, y, 4))

# Evaluation
R2 = r2_score(y, p4(X))
print('Rsquared Value:',R2)

# Visualizing the outcome
# xp is random income generated by numpy between 20000 and 100000 which our model p4 will predict on
xp = np.arange(20000, 100000, 9)
plt.figure(figsize=(11, 5))
plt.scatter(X,y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

###  Findings
A 4th degree polynomial seems to be a good fit for this dataset, It still suggests that my expenses will go up  as my income  But its predictions is better than the linear regression. Expenses flunctuates as income increases so you can say that `The relationship between my income and expenses is a polynomial regression`

```python
n = 4
p = .78
x = np.arange(0, 5)
binomial = stats.binom.pmf(x, n, p)
plt.figure(figsize=(7, 5))
plt.plot(x, binomial, marker='o', c='g')
plt.title("Binomial Distribution of Monthly Income >= Sixty Thousand Naira")
plt.xlabel('Months')
plt.ylabel('Probality of Income');
```
